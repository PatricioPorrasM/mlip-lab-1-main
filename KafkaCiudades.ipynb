### Producer Mode -> Writes Data to Broker

```python
# Create a producer to write data to kafka
# Ref: https://kafka-python.readthedocs.io/en/master/apidoc/KafkaProducer.html

producer = KafkaProducer(
    bootstrap_servers=["localhost:9092"],
    value_serializer=lambda x: x.encode('utf-8')
)
```

```python
print("Writing to Kafka Broker")
with open("movies.csv", newline="", encoding="utf-8") as f:
    next(f)  # skip header
    for line in f:
        line = line.strip()
        if line:
            print(f"Writing line: {line}")
            producer.send(topic=topic, value=line)

producer.flush()
print(f"Data written to topic: {topic}")
```

```python
import time
import random
from datetime import datetime
from kafka import KafkaProducer

topic = 'lab01-pporras'

producer = KafkaProducer(bootstrap_servers='localhost:9092')

users = ['user001', 'user002', 'user003', 'user004', 'user005',
         'user006', 'user007', 'user008', 'user009', 'user010']

# Weighted: mostly 200, some 404, few 500
status_codes = ['200'] * 80 + ['404'] * 15 + ['500'] * 5

def generate_message():
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    user = random.choice(users)
    status = random.choice(status_codes)
    latency = round(random.uniform(10, 500), 2)  # ms
    msg = f"{timestamp},{user},recommendation request,{status},movie1|movie2|movie3,{latency} ms"
    return msg.encode('utf-8')

print(f"Sending messages to topic '{topic}'... Press Ctrl+C to stop.")
count = 0
while True:
    msg = generate_message()
    producer.send(topic, msg)
    count += 1
    print(f"[{count}] {msg.decode('utf-8')}")
    time.sleep(0.5)  # 2 messages per second
```
